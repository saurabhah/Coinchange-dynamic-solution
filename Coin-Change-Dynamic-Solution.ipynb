```python
#https://www.youtube.com/watch?v=H9bfqozjoqs
def coinChange(target,coins):
    
    dp = [target+1] * (target+1)
    
    #print(dp) # if target is 8 =[8, 8, 8, 8, 8, 8, 8, 8]
    
    dp[0]=0
    
    for a in range(1,target+1):#iteration for target 
        for c in coins:
            #if target - coin value is greater than 0
            if a-c >= 0:
                #target the min from and assign to dp[a]
                dp[a] = min(dp[a],1+dp[a-c])
    
    return dp[target]
    
```

```python
coinChange(7,[1,3,4,5])
```
